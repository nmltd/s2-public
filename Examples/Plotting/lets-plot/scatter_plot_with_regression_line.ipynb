```kotlin
/**
 * Demonstrate how to do scatter plot with regression line.
 * This example is found at www.cookbook-r.com/Graphs/Scatterplots_(ggplot2).
 *
 * @author Haksun Li
 */

%use s2, lets-plot

// construct a standard normal random number generator
val rng = NormalRNG(
    0.0, // mean
    1.0) // standard deviation
rng.seed(1234567890L)

// the number of data to generate
val N = 20
// generate the random data to plot
val data = mapOf<String, List<*>>(
    "cond" to List(N / 2) { "A" } + List(N / 2) { "B" },
    "xvar" to List(N) { i:Int-> i }, 
    "yvar" to List(N) { i:Int-> i + rng.nextDouble() * 3 }
)

// construct a scatter plot with regression line
val plot = ggplot(data) { x = "xvar"; y = "yvar" } + // supply data; set legends
    ggsize(300, 250) + // plot size
    geom_point(shape = 1) + // plot the data points
    // draw a scatter plot with regression line
    geom_smooth(se = false) // set se to false to hide the standard error band
plot
```
