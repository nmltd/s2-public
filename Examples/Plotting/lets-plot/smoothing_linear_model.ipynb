```kotlin
/**
 * Demonstrate how to fit data to a linear regression.
 *
 * @author Haksun Li
 */

%use lets-plot, krangl

// Load mpg data
val mpg_df = DataFrame.readCSV("./../../../resources/mpg.csv")
val data = mpg_df.toMap()

// Fit a linear regression to the data
val plot = lets_plot(data) {x="displ"; y="hwy"} + // supply data; set legends
    geom_point() + // plot the data points
    geom_smooth() // The default smoothing method is 'linear model' (or 'lm').
plot
```
