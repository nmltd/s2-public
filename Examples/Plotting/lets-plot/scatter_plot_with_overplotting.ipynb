```kotlin
/**
 * Demonstrate how to do scatter plot with overplotting.
 * This example is found at www.cookbook-r.com/Graphs/Scatterplots_(ggplot2).
 *
 * @author Haksun Li
 */

%use s2, lets-plot

// construct a standard normal random number generator
val rng = NormalRNG(
    0.0, // mean
    1.0) // standard deviation
rng.seed(1234567890L)

// the number of data to generate
val N = 20
// generate the random data to plot
val data = mapOf<String, List<*>>(
    "cond" to List(N / 2) { "A" } + List(N / 2) { "B" },
    "xvar" to List(N) { i:Int-> i }, 
    "yvar" to List(N) { i:Int-> i + rng.nextDouble() * 3 }
)

// Create data with overlapping points.
val data1 = mapOf(
        "xvar" to (data["xvar"] as List<Double>).map { (it / 5).toInt() * 5 },
        "yvar" to (data["yvar"] as List<Double>).map { (it / 5).toInt() * 5 },
)

// construct a scatter plot
val p2 = ggplot(data1) { x = "xvar"; y = "yvar"} + // supply data; set legends
    ggsize(500, 250) + // plot size
    scale_x_continuous(breaks = listOf(0, 5, 10, 15)) + // specify tick values on the x-axis
    // draw a scatter plot
    geom_point(shape = 1, position = position_jitter(width=.1, height=.1)) // `jitter` points to show overplotting
p2
```
