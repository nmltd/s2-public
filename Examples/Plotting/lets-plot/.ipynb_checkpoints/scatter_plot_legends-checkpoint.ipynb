```kotlin
/**
 * Demonstrate how to draw a scatter plot.
 *
 * @author Haksun Li
 */

%use lets-plot, krangl

// download the mpg data
val mpg_df = DataFrame.readCSV("./../../../resources/mpg.csv")
mpg_df.head()
val mpg = mpg_df.toMap()

// Set color to city MPG and shape to drive-train
val plot = ggplot(mpg) {x="displ"; y="hwy"} + // supply data; set legends
    xlab("Engine displacement (L)") + ylab("Highway MPG") + // set axis labels
    ggsize(700, 350) + // plot size
    theme().legendPosition(1, 1).legendJustification(1, 1).legendDirection_horizontal()
    .axisLineY_blank() + // remove vertical axis line
    scale_color_continuous(name="City MPG", 
                           low="dark_blue", high="light_blue", 
                           guide=guide_colorbar(barHeight=10, barWidth=300)) + // set colors
    scale_shape(name="Drive-train", breaks=listOf('f', 'r', '4'), labels=listOf("front", "rear", "4X4")) + // set shapes
    geom_point(size=5.0) {color="cty"; shape="drv"} // draw a scatter plot
plot
```
