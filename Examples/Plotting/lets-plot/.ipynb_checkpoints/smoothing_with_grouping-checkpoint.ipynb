```kotlin
/**
 * Demonstrate how to fit data to a LOESS model with grouping.
 *
 * @author Haksun Li
 */

%use lets-plot, krangl

// Load mpg data
val mpg_df = DataFrame.readCSV("./../../../resources/mpg.csv")
val data = mpg_df.toMap()

/**
 * Let's map the vehicle drive-train type (variable 'drv') to the color of points.
 * This makes it easy to see that points with the same type of the drive-train
 * are forming some kind of groups or clusters.
 * Then fit each group of data with LOESS model.
 */
val plot = lets_plot(data) {x="displ"; y="hwy"} + // supply data; set legends
    geom_point {color="drv"} + // plot the data points
    stat_smooth(method="loess", size=1.0) {color="drv"} // plot the LOESS curves
plot
```
