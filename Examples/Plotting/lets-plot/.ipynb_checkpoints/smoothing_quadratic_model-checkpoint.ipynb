```kotlin
/**
 * Demonstrate how to fit data to a quadratic polinomial regression.
 *
 * @author Haksun Li
 */

%use lets-plot, krangl

// Load mpg data
val mpg_df = DataFrame.readCSV("./../../../resources/mpg.csv")
val data = mpg_df.toMap()

// Fit a 2nd degree polinomial regression to the data groups
val plot = lets_plot(data) {x="displ"; y="hwy"} + // supply data; set legends
    geom_point {color="drv"} + // plot the points
    stat_smooth(method="lm", deg=2, size=1.0) {color="drv"} // plot the regression line
plot
```
