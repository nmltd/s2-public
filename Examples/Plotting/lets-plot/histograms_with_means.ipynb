```kotlin
/**
 * Demonstrate how to plot a histogram overlaid with the kernel density curve.
 *
 * @author Haksun Li
 */

%use s2, lets-plot

// construct a standard normal random number generator
val rng = NormalRNG(
    0.0, // mean
    1.0) // standard deviation
rng.seed(1234567890L)

// the number of data to generate
val N = 200
// generate the random data to plot
val data = mapOf<String, Any>(
    "cond" to List(N) { "A" } + List(N) { "B" },
    "rating" to List(N) { rng.nextDouble() } + List(N) { rng.nextDouble() * 1.5 + 1.5 },
)

// construct a histogram overlaid with kernel density curve
val plot = lets_plot(data) { x = "rating" } + // supply data; set legends
    ggsize(500, 250) + // plot size
    geom_histogram(binWidth=0.5, color="black", fill="white") { y = "..density.." } + // y-axis is density instead of count
    geom_density(alpha=0.2, fill=0xFF6666) // overlay with a transparent density plot
plot
```
