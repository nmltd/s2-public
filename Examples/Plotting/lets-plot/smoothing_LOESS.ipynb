```kotlin
/**
 * Demonstrate how to fit data to a LOESS model.
 *
 * @author Haksun Li
 */

%use lets-plot, krangl

// Load mpg data
val mpg_df = DataFrame.readCSV("./../../../resources/mpg.csv")
val data = mpg_df.toMap()

// Fit the data with a LOESS model.
val plot = lets_plot(data) {x="displ"; y="hwy"} + // supply data; set legends
    geom_point() + // plot the data points
    stat_smooth(method="loess", size=1.0) // plot the LOESS curve
plot
```
