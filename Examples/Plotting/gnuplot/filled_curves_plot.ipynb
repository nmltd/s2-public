```kotlin
/**
 * Demonstrate how to draw filled curves with shaded regions.
 *
 * @author Haksun Li
 */

%use s2

import java.net.URL

val p = Gnuplot()
p.setTitle("error on y represented by the shaded region")
p.getXAxis().setLabel("Time (sec)")
p.getYAxis().setLabel("Rate")
p.getYAxis().setLogScale(true)
p.set("grid")

// add error bars
val data = TextDataSet(BufferedReader(InputStreamReader(URL("https://raw.githubusercontent.com/nmltd/s2-public/main/Data/Plotting/silver.dat").openStream())))
val dp1 = DataSetPlot(data)
dp1.usingColumns("1:($2+$3):($2-$3)") // last two columns are error range
dp1.setTitle("shaded error range")
dp1.getPlotStyle().setStyle(Style.FILLEDCURVES) // shaded region
dp1.getPlotStyle().setColor(NamedPlotColor.LIGHT_GRAY)
p.addPlot(dp1)

// add error bars
val dp2 = DataSetPlot(data)
dp2.usingColumns("1:2")
dp2.setTitle("y")
dp2.getPlotStyle().setStyle(Style.LINES) // line plot
p.addPlot(dp2)

p.plot()
```
