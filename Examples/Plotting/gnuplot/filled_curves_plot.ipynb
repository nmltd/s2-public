```kotlin
/**
 * Demonstrate how to draw filled curves with shaded regions.
 *
 * @author Haksun Li
 */

%use s2

import java.net.URL
import java.io.BufferedReader
import java.io.InputStreamReader

val p = JGnuplot()
p.setTitle("error on y represented by the shaded region")
p.getXAxis().setLabel("Time (sec)")
p.getYAxis().setLabel("Rate")
p.getYAxis().setLogScale(true)
p.set("grid")

// add error bars
val dp1 = DataSetPlot(FileDataSet(BufferedReader(InputStreamReader(URL("https://s2-nmdev.s3.eu-west-3.amazonaws.com/resources/silver.dat").openStream()))))
dp1.setUsingColumns("1:($2+$3):($2-$3)") // last two columns are error range
dp1.setTitle("shaded error range")
dp1.getPlotStyle().setStyle(Style.FILLEDCURVES) // shaded region
dp1.getPlotStyle().setColor(NamedPlotColor.LIGHT_GRAY)
p.addPlot(dp1)

// add error bars
val dp2 = DataSetPlot()
dp2.setUsingColumns("1:2")
dp2.setTitle("y")
dp2.getPlotStyle().setStyle(Style.LINES) // line plot
p.addPlot(dp2)

p.plot()
```
