```kotlin
%use s2

import java.net.URL
import java.io.BufferedReader
import java.io.InputStreamReader

val p = JGnuplot()
p.setTitle("Monotonic cubic splines")
p.getXAxis().setBoundaries(233.0, 388.0)
p.getYAxis().setBoundaries(1.75, 32.0)

// Raw data
val dp1 = DataSetPlot(FileDataSet(BufferedReader(InputStreamReader(URL("https://s2-nmdev.s3.eu-west-3.amazonaws.com/resources/silver.dat").openStream()))))
dp1.setUsingColumns("1:2")
dp1.setTitle("raw data")
dp1.getPlotStyle().setPointType(7)
dp1.getPlotStyle().setColor(NamedPlotColor.LIGHT_GRAY)
p.addPlot(dp1)

// Csplines
val dp2 = DataSetPlot()
dp2.setUsingColumns("1:2")
dp2.setTitle("smooth cspline")
dp2.setSmooth(Smooth.CSPLINES)
dp2.getPlotStyle().setColor(NamedPlotColor.RED)
p.addPlot(dp2)

// Monotonic csplines
val dp3 = DataSetPlot()
dp3.setUsingColumns("1:2")
dp3.setTitle("smooth mcs")
dp3.setSmooth(Smooth.MCSPLINES)
dp3.getPlotStyle().setColor(NamedPlotColor.BLUE)
p.addPlot(dp3)

p.plot()
```
