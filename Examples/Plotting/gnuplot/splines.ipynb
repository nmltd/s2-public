```kotlin
%use s2

import java.net.URL

val p = Gnuplot()
p.setTitle("Monotonic cubic splines")
p.getXAxis().setBoundaries(233.0, 388.0)
p.getYAxis().setBoundaries(1.75, 32.0)

// Raw data
val data = TextDataSet(BufferedReader(InputStreamReader(URL("https://raw.githubusercontent.com/nmltd/s2-public/main/Data/Plotting/silver.dat").openStream())))
val dp1 = DataSetPlot(data)
dp1.usingColumns("1:2")
dp1.setTitle("raw data")
dp1.getPlotStyle().setPointType(7)
dp1.getPlotStyle().setColor(NamedPlotColor.LIGHT_GRAY)
p.addPlot(dp1)

// Csplines
val dp2 = DataSetPlot(data)
dp2.usingColumns("1:2")
dp2.setTitle("smooth cspline")
dp2.setSmooth(Smooth.CSPLINES)
dp2.getPlotStyle().setColor(NamedPlotColor.RED)
p.addPlot(dp2)

// Monotonic csplines
val dp3 = DataSetPlot(data)
dp3.usingColumns("1:2")
dp3.setTitle("smooth mcs")
dp3.setSmooth(Smooth.MCSPLINES)
dp3.getPlotStyle().setColor(NamedPlotColor.BLUE)
p.addPlot(dp3)

p.plot()
```
