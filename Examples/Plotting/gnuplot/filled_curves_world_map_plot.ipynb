```kotlin
/**
 * Demonstrate how to draw filled curves on a world map.
 *
 * @author Haksun Li
 */

%use s2

import java.net.URL
import java.io.BufferedReader
import java.io.InputStreamReader

val p = JGnuplot()
p.setTitle("World Map with Filled Curves")
p.setTitleFontSize(20)
p.getXAxis().setTicLabelFormat("")
p.getYAxis().setTicLabelFormat("")
p.setGrid(0.5)
p.getXAxis().setBoundaries(-180.0, 180.0)
p.getYAxis().setBoundaries(-70.0, 80.0)

val dataSetPlot = DataSetPlot(FileDataSet(BufferedReader(InputStreamReader(URL("https://s2-nmdev.s3.eu-west-3.amazonaws.com/resources/world.dat").openStream()))))
dataSetPlot.hideTitle()
val style = dataSetPlot.getPlotStyle()
style.setStyle(Style.FILLEDCURVES)
// val fillStyle = FillStyle(Fill.SOLID)
// fillStyle.setDensity(1.0)
// style.setFill(fillStyle)
style.setColor(NamedPlotColor.DARK_GOLDENROD)
p.addPlot(dataSetPlot)

p.plot()
```
