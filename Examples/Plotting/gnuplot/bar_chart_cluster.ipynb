```kotlin
/**
 * Demonstrate how to plot bar charts, cluster or column style.
 *
 * @author Haksun Li
 */

%use s2

import java.net.URL
import java.io.BufferedReader
import java.io.InputStreamReader

val p = JGnuplot()
p.setTitle("Bar Chart")
p.setTitleFontSize(20)
// set the default style of data plot to histogram
p.setDefaultStyle("data", "histogram")
// set the default style of histograms to cluster with gap of 1.0
p.setDefaultStyle("histogram", "cluster gap 1")
// set the default style of filling to solid with border whose linetype is -1
p.setDefaultStyle("fill", "solid border -1")
p.getYAxis().setBoundaries(0.0, 300000.0)
p.setBoxWidth(0.9)
// rotate the x tics
p.set("xtic", "rotate by -45 scale 0")

// download the data from the Internet
val dp1 = DataSetPlot(FileDataSet(BufferedReader(InputStreamReader(URL("https://s2-nmdev.s3.eu-west-3.amazonaws.com/resources/immigration.dat").openStream()))))
dp1.setUsingColumns("6:xtic(1)")
p.addPlot(dp1)

// use the same data set
val dp2 = DataSetPlot()
dp2.setUsingColumns("12")
p.addPlot(dp2)

val dp3 = DataSetPlot()
dp3.setUsingColumns("13")
p.addPlot(dp3)

val dp4 = DataSetPlot()
dp4.setUsingColumns("14")
p.addPlot(dp4)

p.plot()
```
