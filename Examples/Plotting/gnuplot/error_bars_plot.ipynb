```kotlin
/**
 * Demonstrate how to draw error bars.
 *
 * An errorbar plot usually requires three columns. The third column is 'error'.
 * For more information, refer to "http://gnuplot.sourceforge.net/docs_4.2/node140.html".
 *
 * @author Haksun Li
 */

%use s2

import java.net.URL

val p = Gnuplot()
p.setTitle("Error Bars")
p.setTitleFontSize(20)

// add error bars
val data = TextDataSet(BufferedReader(InputStreamReader(URL("https://raw.githubusercontent.com/nmltd/s2-public/main/Data/Plotting/battery.dat").openStream())))
val dp = DataSetPlot(data)
dp.usingColumns("1:2:3")
dp.setTitle("Power")
dp.getPlotStyle().setStyle(Style.XERRORBARS) // set the style to XERRORBARS
p.addPlot(dp)

// add theory values
p.addPlot("1.53**2*x/(5.67+x)**2")
p.getPlots().get(1).setTitle("Theory")

p.plot()
```
