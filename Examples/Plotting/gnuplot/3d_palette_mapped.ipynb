```kotlin
/**
 * Demonstrate how to draw palette-mapped 3d and 4d data as color/gray maps and surfaces.
 * "pm3d" is an splot style.
 *
 * @author Haksun Li
 */

%use s2

val p = JGnuplot(true) // true: 3d plot
p.setTitle("pm3d demo: radial sinc function")
p.setLegend(Legend.OFF)
p.set("pm3d")
p.set("palette")
p.set("border", "4095")

p.getAxis("x").setBoundaries(-15.0, 15.0)
p.getAxis("y").setBoundaries(-15.0, 15.0)
p.getAxis("z").setBoundaries(-0.25, 1.0)

p.addPlot("sin(sqrt(x**2+y**2))/sqrt(x**2+y**2)")

p.plot()
```
