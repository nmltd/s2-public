```kotlin
/**
 * Demonstrate how to draw filled curves with shaded regions.
 *
 * @author Haksun Li
 */

%use s2

val p = JGnuplot()
p.setLegend(Legend.OUTSIDE)
p.setTitle("Filled Curves")
p.setTitleFontSize(20)
p.addPlot("x**2") // x^2
val style1 = p.getPlots().get(0).getPlotStyle()
style1.setStyle(Style.FILLEDCURVES)
style1.setStyleOptions("x2") // fill to the top side (the x2 axis)
p.addPlot("50 - x**2") // 50 - x^2
val style2 = p.getPlots().get(1).getPlotStyle()
style2.setStyle(Style.FILLEDCURVES)
style2.setStyleOptions("x1") // fill to the bottom side (the x1 axis)
p.addPlot("x**2") // x^2

p.plot()
```
