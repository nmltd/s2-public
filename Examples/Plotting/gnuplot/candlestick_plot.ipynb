```kotlin
/**
 * Demonstrate how to draw candle sticks.
 *
 * @author Haksun Li
 */

%use s2

import java.net.URL
import java.io.BufferedReader
import java.io.InputStreamReader

val p = JGnuplot()
p.setTitle("Candle Sticks")
p.setTitleFontSize(20)
p.setLegend(Legend.OFF)
p.setBoxWidth(0.2)
p.getXAxis().setBoundaries(0.0, 11.0)
p.getYAxis().setBoundaries(0.0, 10.0)

// download the data from the Internet
val dataSetPlot = DataSetPlot(FileDataSet(BufferedReader(InputStreamReader(URL("https://s2-nmdev.s3.eu-west-3.amazonaws.com/resources/candlesticks.dat").openStream()))))
// Candles tick plot requires five columns of data.
dataSetPlot.setUsingColumns("1:3:2:6:5")
// set the style of data plot to candle sticks
dataSetPlot.getPlotStyle().setStyle(Style.CANDLESTICKS)
p.addPlot(dataSetPlot)

p.plot()
```
