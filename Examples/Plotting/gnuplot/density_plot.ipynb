```kotlin
%use s2

val rand = java.util.Random(123)
val n = 200
val values = mapOf<String, List<Double>>(
    "rating" to List(n) { rand.nextGaussian() } + List(n) { rand.nextGaussian() * 1.5 + 1.5 },
)

val p = Gnuplot()
val dp1 = DensityPlot(MapDataSet(values))
dp1.usingColumns("'rating'")
dp1.setTitle("density")
p.addPlot(dp1)

p.plot()
```
