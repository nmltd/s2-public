```kotlin
/**
 * Demonstrate how to plot bar charts, row-stacked style.
 *
 * @author Haksun Li
 */

%use s2

import java.net.URL

val p = Gnuplot()
p.setTitle("Bar Chart Stacked")
p.setTitleFontSize(20)
// set the default style of data plot to histogram
p.setDefaultStyle("data", "histogram")
// set the default style of histograms to rowstacked
p.setDefaultStyle("histogram", "rowstacked")
// set the default style of filling to solid with border whose linetype is -1
p.setDefaultStyle("fill", "solid border -1")
p.setBoxWidth(0.9)
// rotate the x tics
p.set("xtic", "rotate by -45 scale 0")

// download the data from the Internet
val data = TextDataSet(BufferedReader(InputStreamReader(URL("https://raw.githubusercontent.com/nmltd/s2-public/main/Data/Plotting/immigration.dat").openStream())))
val dp1 = DataSetPlot(data)
dp1.usingColumns("6:xtic(1)")
p.addPlot(dp1)

// use the same data set
val dp2 = DataSetPlot(data)
dp2.usingColumns("12")
p.addPlot(dp2)

val dp3 = DataSetPlot(data)
dp3.usingColumns("13")
p.addPlot(dp3)

val dp4 = DataSetPlot(data)
dp4.usingColumns("14")
p.addPlot(dp4)

p.plot()
```
