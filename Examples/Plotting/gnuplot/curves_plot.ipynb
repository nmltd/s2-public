```kotlin
/**
 * Demonstrate how to draw curves.
 *
 * @author Haksun Li
 */

%use s2

val p = JGnuplot()
p.setTitle("simple function plots")
p.setTitleFontSize(20)
p.setLegend(Legend.LEFT)
// p.setSamples(50) // set the sample rate when plotting functions
p.addPlot("sin(x)")
p.addPlot("atan(x)")
p.addPlot("cos(atan(x))")
p.plot()
```
