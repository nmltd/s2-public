```kotlin
/**
 * Demonstrate how to draw box plot.
 *
 * @author Haksun Li
 */

%use s2

import java.net.URL

val p = Gnuplot()
p.setTitle("distribution of energy usage of the continents, grouped by energy source type")
p.setLegend(Legend.OFF)
p.getYAxis().setLabel("Billion Tons of Oil Equivalent")
// set the default style of data plot to boxplot
p.setDefaultStyle("data", "boxplot")
// set the default style of filling to solid with border whose linetype is -1
p.setDefaultStyle("fill", "solid 0.5 border -1")
p.setBoxWidth(0.5)
p.setPointSize(0.5)

// download the data from the Internet
val data = TextDataSet(BufferedReader(InputStreamReader(URL("https://raw.githubusercontent.com/nmltd/s2-public/main/Data/Plotting/energy_circles.dat").openStream())))
val dp = DataSetPlot(data)
dp.usingColumns("(1):($8/1.e6):(0):4")
p.addPlot(dp)

p.plot()
```
