```kotlin
%use s2, plotly

// data to plot
val x1 = listOf(1, 2, 3, 4)
val y1 = listOf(2, 3, 5, 8)

// error data
val err = listOf(0.5, 0.75, 1.0, 1.25)

// construct the traces of the scatter plot
val BothErr = Scatter {
    x.set(x1) // set data for x axis
    y.set(y1) // set data for y axis
    name = "both errors"
    error_y {  // set this property to plot errors
        type = ErrorType.data
        array = err // set data for error
        visible = true
    }
}

val PosErr = Scatter {
    x.set(x1) // set data for x axis
    y.set(y1.map { it + 4 }.toList()) // modify and set data for y axis
    name = "positive err"
    error_y {  // set this property to plot errors
        type = ErrorType.data
        array = err // set data for error
        symmetric = false
        visible = true
    }
}

val NegErr = Scatter {
    x.set(x1) // set data for x axis
    y.set(y1.map { it + 8 }.toList()) // modify and set data for y axis
    name = "negative err"
    error_y {  // set this property to plot errors
        type = ErrorType.data
        arrayminus = err // set data for error
        symmetric = false
        visible = true
    }
}

// plot the traces
val plot = Plotly.plot {
    traces(BothErr, PosErr, NegErr) // supply traces

    layout { // set layout properties
        title = "Asymmetric Error Bars"
        legend { // set legend properties
            x = 0.05
            borderwidth = 1
            xanchor = XAnchor.left
            yanchor = YAnchor.top
            traceorder = TraceOrder.reversed
        }
    }
}

// display the plot
plot
```
