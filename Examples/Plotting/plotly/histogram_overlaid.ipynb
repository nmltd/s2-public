```kotlin
%use s2, plotly

// construct a standard normal random number generator
val rng = NormalRNG(0.0,0.1)
rng.seed(1234567890L)

// the number of data to generate
val N = 500

// generate the random data to plot
val x1 = List(N) { rng.nextDouble() + 1}
val x2 = List(N) { rng.nextDouble() + 1.2}

// construct the traces of the histogram
val hist1 = Histogram {
    x.set(x1) // set data
    opacity = 0.5 // set opacity to view overlaid histogram
    marker { // set marker properties
        color("Blue")
    }
    name = "Random data 1"
}

val hist2 = Histogram {
    x.set(x2) // set data
    opacity = 0.6 // set opacity to view overlaid histogram
    marker { // set marker properties
        color("Pink")
    }
    name = "Random data 2"
}

// plot the traces
val plot = Plotly.plot {
    traces(hist1,hist2) // supply traces
    
    layout { // set layout properties
        bargap = 0.1
        barmode = BarMode.overlay // set barmode to overlay
        title {
            text = "Overlaid Histogram"
            font {
                size = 20
                color("black")
            }
        }
        // set axes properties
        xaxis {
            title {
                text = "Value"
                font {
                    size = 16
                }
            }
        }
        yaxis {
            title {
                text = "Count"
                font {
                    size = 16
                }
            }
        }
    }
}

// display the plot
plot
```
