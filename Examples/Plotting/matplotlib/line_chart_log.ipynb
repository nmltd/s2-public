```kotlin
%use s2

@file:Repository("https://github.com/sh0nk/matplotlib4j")
@file:DependsOn("com.github.sh0nk:matplotlib4j:0.5.0")
import com.github.sh0nk.matplotlib4j.*
import com.github.sh0nk.matplotlib4j.builder.*

// construct a standard normal random number generator
val rng = NormalRNG(0.0, 1.0)
rng.seed(1234567890L)

// the number of data to generate
val N = 200

// generate the random data to plot
val x = List(N) { rng.nextDouble() * log10(4.0) }

// generate the plot
val plt = Plot.create()
plt.plot().add(x) // set data
plt.title("Logarithmic Line Chart") // set title
plt.xscale(ScaleBuilder.Scale.log) // set scale to log axis
plt.xlabel("Frequency") // set label

// display the plot
// plt.show()
plt.savefig("/home/jovyan/tmp/line_chart_log.png")
plt.executeSilently()
```
