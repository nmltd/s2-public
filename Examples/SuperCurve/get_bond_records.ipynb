```kotlin
/**
 * Retrieve the listed treasury bonds for a date.
 *
 * @author Haksun Li
 */

%use s2, krangl

// the date of interest
val date = LocalDate.of(2019, 9, 2)

// get the bonds for the date
val records = BondSnapshotDB.getBonds(
    dev.nm.s2.app.supercurve.DBConnDefault(), // database connection
    date, // the date
    Issuer.TB.names() // TB = treasury bonds
)

// convert `records` to a data frame
val headers = BondRecord.Field.values().map { it.toString() }
val values: MutableList<Any> = ArrayList()
records.forEach { r ->
    headers.forEach { f ->
        values.add( BondRecord.getString(r, f) ?: "" )
    }
}

// print out the bonds
val df = dataFrameOf(headers)(values)
df.print(title="Bond Records", maxRows=records.size(), rowNumbers=true)
```
