```kotlin
// the cutoff point for x
fun x_cutoff(a : Double, b : Double, c : Double, d : Double) : Double {
    var x_cutoff = d + b;
    x_cutoff /= a - b - c + d;

    return x_cutoff;
}
```

```kotlin
// the cutoff point for y
fun y(a : Double, b : Double, c : Double, d : Double, x : Double) : Double {
    var y_cutoff = (d - c) * x - d;
    y_cutoff /= (a - b - c + d) * x - (d - b);

    return y_cutoff;
}
```

```kotlin
fun y1(x : Double) : Double {
    val a = 3.0;
    val b = -2.0;
    val c = -2.0;
    val d = 1.0;

    return y(a, b, c, d, x);
}
```

```kotlin
// the upper and lower bounds for y
fun y_interval(a : Double, b : Double, c : Double, d : Double) : Pair<Double, Double> {
    
    var yl = y(a, b, c, d, 0.0); // y lower bound for x = 0
    var yu = y(a, b, c, d, 1.0); // y upper  bound for x = 1

    var p : Pair<Double, Double> = Pair(yl, yu);
    return p;
}
```

```kotlin
fun y1(x : Double) : Double {
    val a = 3.0;
    val b = -2.0;
    val c = -2.0;
    val d = 1.0;

    return y(a, b, c, d, x);
}
```

```kotlin
val x_values = (0..100).map{ it*0.01 } // x from 0.0 to 1.0 with an increment of 0.01
```

```kotlin
val y1_values =  x_values.map{y1(it)} // the y boundary values
```

```kotlin
%use plotly

Plotly.plot {
    scatter {
        x.set(x_values)
        y.set(y1_values)
    }
    
    layout{
        title = "the y probability"
        xaxis {
            title = "x"
            range = 0.0..1.0
        }
        yaxis {
            title = "y"
            range = 0.0..1.0
        }
    }
}
```

```kotlin
fun plot_y(a : Double, b : Double, c : Double, d : Double) : Plot {
    val x_values = (0..100).map{ it*0.01 } // x from 0.0 to 1.0 with an increment of 0.01 
    val y_values =  x_values.map{y(a, b, c, d, it)} // the y boundary values
    
    val y_interval = y_interval(a, b, c, d);
    println("y interval = $y_interval");
    
    val plot = Plotly.plot {
        scatter {
            x.set(x_values)
            y.set(y_values)
        }
    
        layout{
            title = "the y probability"
            xaxis {
                title = "x"
                range = 0.0..1.0
            }
            yaxis {
                title = "y"
                range = 0.0..1.0
            }
        }
    }
    
    return plot;
}
```

```kotlin
val plot1 = plot_y(3.0, -2.0, -2.0, 1.0)
plot1
```
